```python
class BiLSTM(nn.Module):
        def __init__(self, input_size, hidden_size_1, hidden_size_2, num_layers, output_size):
            super(BiLSTM, self).__init__()
            self.hidden_size_1 = hidden_size_1
            self.hidden_size_2 = hidden_size_2
            self.num_layers = num_layers
            self.bilstm1 = nn.LSTM(input_size, hidden_size_1, num_layers, batch_first=True, bidirectional=True)
            self.bilstm2 = nn.LSTM(hidden_size_1*2, hidden_size_2, num_layers, batch_first=True, bidirectional=True)
            self.fc = nn.Linear(2*hidden_size_2, output_size)

        def forward(self, x):
            h0_1 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size_1, dtype=torch.float32).to(x.device)
            c0_1 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size_1, dtype=torch.float32).to(x.device)
            out, _ = self.bilstm1(x, (h0_1, c0_1))
            # print(out.shape)
            h0_2 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size_2, dtype=torch.float32).to(x.device)
            c0_2 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size_2, dtype=torch.float32).to(x.device)
            out, _ = self.bilstm2(out, (h0_2, c0_2))
            # print(out.shape)
            out = self.fc(out[:, -1, :])
            return out
```

```python
class SVM:
    def __init__(self, x_train, y_train, x_test, y_test, k):
        self.svm = svm.SVC(kernel='linear')
        self.x_train = x_train
        self.y_train = y_train
        self.x_test = x_test
        self.y_test = y_test
        self.k = k
        #print(self.x_train.shape, self.y_train.shape, self.x_test.shape, self.y_test.shape)

    def feature_selection_ExtraTreesClassifier(self):
        clf = ExtraTreesClassifier(n_estimators=100, random_state=42)
        clf.fit(self.x_train, self.y_train)
        importances = clf.feature_importances_
        indices = np.argsort(importances)[::-1]
        self.x_train = self.x_train[:, indices[:self.k]]
        self.x_test = self.x_test[:, indices[:self.k]]
    
    def train(self):
        self.svm.fit(self.x_train, self.y_train)

    def pred(self):
        y_pred = self.svm.predict(self.x_test)
        return y_pred
```

```python
class RandomForest:
    def __init__(self, x_train, y_train, x_test, y_test, k):
        self.rf = RandomForestClassifier(n_estimators=100, random_state=42)
        self.k = k
        self.x_train = x_train
        self.y_train = y_train
        self.x_test = x_test
        self.y_test = y_test
        #print(self.x_train.shape, self.y_train.shape, self.x_test.shape, self.y_test.shape)

    def feature_selection_SelectKBest(self):
        k_best = SelectKBest(mutual_info_classif, k=self.k)
        k_best.fit(self.x_train, self.y_train)
        self.x_train = k_best.transform(self.x_train)
        self.x_test = k_best.transform(self.x_test)
        
    def train(self):
        self.rf.fit(self.x_train, self.y_train)

    def pred(self):
        y_pred = self.rf.predict(self.x_test)
        return y_pred
```

```python
class KNN:
    def __init__(self, x_train, y_train, x_test, y_test, n):
        self.knn = KNeighborsClassifier(n_neighbors=n)
        self.x_train = x_train
        self.y_train = y_train
        self.x_test = x_test
        self.y_test = y_test

    def train(self):
        self.knn.fit(self.x_train, self.y_train)

    def pred(self):
        y_pred = self.knn.predict(self.x_test)
        return y_pred
```

##### Ensemble with:
DL model using GloVe word embeddings  
DL model using fastText word embeddings  
SVM with ExtraTreesClassifier algorithm  
Random Forest with SelectKBest algorithm  
kNN (k = 1) with fastText word embeddings

```python
with open("./results/ensembles/performance_ensemble1.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow([column_headings[0], column_headings[1], column_headings[2]])

all_f1_macro_scores = []
all_f1_micro_scores = []

for morbidity in morbidities:
    print(morbidity)
    train_preprocessed_df = pd.read_csv('./dataset/train/train_data_intuitive_preprocessed.csv')
    train_preprocessed_df = train_preprocessed_df[train_preprocessed_df[morbidity].isin([1.0, 0.0])]

    train_df, test_df = train_test_split(train_preprocessed_df, test_size=0.2, random_state=42)

    X_train_glove, Y_train_glove, words_train_glove = GloVeFeatureGeneration(train_df, morbidity).matrix_gen()
    X_test_glove, Y_test_glove, words_test_glove = GloVeFeatureGeneration(test_df, morbidity).matrix_gen()
    # X_train_glove = np.abs(np.average(X_train_glove, axis=1))
    # X_test_glove = np.abs(np.average(X_test_glove, axis=1))
    
    # print(X_train_glove.shape)
    # print(Y_train_glove.shape)

    X_train_glove = torch.tensor(X_train_glove, dtype=torch.float32)
    Y_train_glove = torch.tensor(Y_train_glove, dtype=torch.float32)
    X_test_glove = torch.tensor(X_test_glove, dtype=torch.float32)
    Y_test_glove = torch.tensor(Y_test_glove, dtype=torch.float32)
            
    #Model parameters
    input_size = X_train_glove.size(2)
    hidden_size_1 = 20
    hidden_size_2 = 10
    num_layers = 1
    output_size = 1

    bilstm = BiLSTM(input_size, hidden_size_1, hidden_size_2, num_layers, output_size)

    bilstm.load_state_dict(torch.load(f'./models/Ensembles/pretrained-DL-models/bilstm_model_{morbidity}_glove.pt', map_location=torch.device('cpu')), strict=False)

    with torch.no_grad():
        bilstm.eval()
        y_hat = bilstm(X_test_glove).squeeze().detach().numpy()

    y_pred_DL_glove = []
    for val in y_hat.data:
        if val <= 0.6:
            y_pred_DL_glove.append(0)
        else:
            y_pred_DL_glove.append(1)
            
    y_pred_DL_glove = torch.tensor(y_pred_DL_glove)

    #DL model 2

    X_train_fasttext, Y_train_fasttext, words_train_fasttext = FastTextFeatureGeneration(train_df, morbidity).matrix_gen()
    X_test_fasttext, Y_test_fasttext, words_test_fasttext = FastTextFeatureGeneration(test_df, morbidity).matrix_gen()
    # X_train_fasttext = np.average(X_train_fasttext, axis=1)
    # X_test_fasttext = np.average(X_test_fasttext, axis=1)

    # print(X_train_fasttext.shape)
    # print(Y_train_fasttext.shape)

    X_train_fasttext = torch.tensor(X_train_fasttext, dtype=torch.float32)
    Y_train_fasttext = torch.tensor(Y_train_fasttext, dtype=torch.float32)
    X_test_fasttext = torch.tensor(X_test_fasttext, dtype=torch.float32)
    Y_test_fasttext = torch.tensor(Y_test_fasttext, dtype=torch.float32)
            
    #Model parameters
    input_size = X_train_fasttext.size(2)
    hidden_size_1 = 20
    hidden_size_2 = 10
    num_layers = 1
    output_size = 1

    bilstm = BiLSTM(input_size, hidden_size_1, hidden_size_2, num_layers, output_size)

    bilstm.load_state_dict(torch.load(f'./models/Ensembles/pretrained-DL-models/bilstm_model_{morbidity}_fasttext.pt', map_location=torch.device('cpu')), strict=False)

    with torch.no_grad():
        bilstm.eval()
        y_hat = bilstm(X_test_fasttext).squeeze().detach().numpy()

    y_pred_DL_fasttext = []
    for val in y_hat.data:
        if val <= 0.6:
            y_pred_DL_fasttext.append(0)
        else:
            y_pred_DL_fasttext.append(1)
            
    y_pred_DL_fasttext = torch.tensor(y_pred_DL_fasttext)

    # CML models

    # train a KNN model on the FastText features
    
    X_train_fasttext = np.average(X_train_fasttext, axis=1)
    X_test_fasttext = np.average(X_test_fasttext, axis=1)

    knn_obj = KNN(X_train_fasttext, Y_train_fasttext, X_test_fasttext, Y_test_fasttext, 1)
    knn_obj.train()
    y_pred_knn_fasttext = knn_obj.pred()

    y_preds = [y_pred_DL_glove, y_pred_DL_fasttext, y_pred_knn_fasttext]
    
    # perform majority vote
    majority_vote = [mode([y_pred[i] for y_pred in y_preds]) for i in range(len(y_preds[0]))]

    f1_macro = f1_score(Y_test_fasttext, majority_vote, average='macro')
    f1_micro = f1_score(Y_test_fasttext, majority_vote, average='micro')

    print(f"Macro F1 score: {f1_macro} and Micro F1 Score {f1_micro}")

    row_heading = morbidity

    data = [f1_macro, f1_micro]
    all_f1_macro_scores.append(f1_macro)
    all_f1_micro_scores.append(f1_micro)

    with open("./results/ensembles/performance_ensemble1.csv", "a", newline="") as file:
        writer = csv.writer(file)
        row = [row_heading]
        row.extend(data)
        writer.writerow(row)


with open("./results/ensembles/performance_ensemble1.csv", "a", newline="") as file:
    writer = csv.writer(file)
    row = ["Overall-Average"]
    row.extend([sum(all_f1_macro_scores)/len(all_f1_macro_scores),  sum(all_f1_micro_scores)/len(all_f1_micro_scores) ])
    writer.writerow(row)    
```
